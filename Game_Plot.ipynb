```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import matplotlib.patches as mpatches

print("=" * 80)
print("POPULAR GAMES VISUALIZATION (2015-2025)")
print("=" * 80)

# Creating comprehensive gaming dataset
def create_games_dataset():
 
    print("\n Creating gaming dataset...")
    
    games_data = [
        # 2015
        {'Year': 2015, 'Game': 'The Witcher 3', 'Sales_Million': 40, 'Genre': 'RPG', 'Platform': 'Multi', 'Rating': 9.3},
        {'Year': 2015, 'Game': 'Fallout 4', 'Sales_Million': 25, 'Genre': 'RPG', 'Platform': 'Multi', 'Rating': 8.7},
        {'Year': 2015, 'Game': 'GTA V', 'Sales_Million': 60, 'Genre': 'Action', 'Platform': 'Multi', 'Rating': 9.5},
        {'Year': 2015, 'Game': 'Rocket League', 'Sales_Million': 20, 'Genre': 'Sports', 'Platform': 'Multi', 'Rating': 8.5},
        
        # 2016
        {'Year': 2016, 'Game': 'Overwatch', 'Sales_Million': 50, 'Genre': 'Shooter', 'Platform': 'Multi', 'Rating': 9.1},
        {'Year': 2016, 'Game': 'Pokémon GO', 'Sales_Million': 35, 'Genre': 'Mobile', 'Platform': 'Mobile', 'Rating': 7.8},
        {'Year': 2016, 'Game': 'Uncharted 4', 'Sales_Million': 16, 'Genre': 'Action', 'Platform': 'PS4', 'Rating': 9.3},
        {'Year': 2016, 'Game': 'Dark Souls III', 'Sales_Million': 10, 'Genre': 'RPG', 'Platform': 'Multi', 'Rating': 8.9},
        
        # 2017
        {'Year': 2017, 'Game': 'PUBG', 'Sales_Million': 75, 'Genre': 'Battle Royale', 'Platform': 'Multi', 'Rating': 8.2},
        {'Year': 2017, 'Game': 'Zelda: BOTW', 'Sales_Million': 31, 'Genre': 'Adventure', 'Platform': 'Switch', 'Rating': 9.7},
        {'Year': 2017, 'Game': 'Horizon Zero Dawn', 'Sales_Million': 20, 'Genre': 'Action', 'Platform': 'PS4', 'Rating': 8.9},
        {'Year': 2017, 'Game': 'Destiny 2', 'Sales_Million': 25, 'Genre': 'Shooter', 'Platform': 'Multi', 'Rating': 8.0},
        
        # 2018
        {'Year': 2018, 'Game': 'Fortnite', 'Sales_Million': 125, 'Genre': 'Battle Royale', 'Platform': 'Multi', 'Rating': 8.3},
        {'Year': 2018, 'Game': 'Red Dead 2', 'Sales_Million': 55, 'Genre': 'Action', 'Platform': 'Multi', 'Rating': 9.7},
        {'Year': 2018, 'Game': 'God of War', 'Sales_Million': 23, 'Genre': 'Action', 'Platform': 'PS4', 'Rating': 9.5},
        {'Year': 2018, 'Game': 'Spider-Man', 'Sales_Million': 33, 'Genre': 'Action', 'Platform': 'PS4', 'Rating': 8.7},
        
        # 2019
        {'Year': 2019, 'Game': 'Apex Legends', 'Sales_Million': 45, 'Genre': 'Battle Royale', 'Platform': 'Multi', 'Rating': 8.5},
        {'Year': 2019, 'Game': 'Sekiro', 'Sales_Million': 10, 'Genre': 'Action', 'Platform': 'Multi', 'Rating': 9.0},
        {'Year': 2019, 'Game': 'Death Stranding', 'Sales_Million': 10, 'Genre': 'Action', 'Platform': 'PS4', 'Rating': 8.2},
        {'Year': 2019, 'Game': 'Call of Duty MW', 'Sales_Million': 30, 'Genre': 'Shooter', 'Platform': 'Multi', 'Rating': 8.1},
        
        # 2020
        {'Year': 2020, 'Game': 'Animal Crossing NH', 'Sales_Million': 45, 'Genre': 'Simulation', 'Platform': 'Switch', 'Rating': 9.0},
        {'Year': 2020, 'Game': 'The Last of Us 2', 'Sales_Million': 10, 'Genre': 'Action', 'Platform': 'PS4', 'Rating': 9.3},
        {'Year': 2020, 'Game': 'Cyberpunk 2077', 'Sales_Million': 25, 'Genre': 'RPG', 'Platform': 'Multi', 'Rating': 7.8},
        {'Year': 2020, 'Game': 'Among Us', 'Sales_Million': 60, 'Genre': 'Social', 'Platform': 'Multi', 'Rating': 8.0},
        
        # 2021
        {'Year': 2021, 'Game': 'Resident Evil Village', 'Sales_Million': 8, 'Genre': 'Horror', 'Platform': 'Multi', 'Rating': 8.5},
        {'Year': 2021, 'Game': 'It Takes Two', 'Sales_Million': 10, 'Genre': 'Adventure', 'Platform': 'Multi', 'Rating': 8.9},
        {'Year': 2021, 'Game': 'Ratchet & Clank', 'Sales_Million': 4, 'Genre': 'Action', 'Platform': 'PS5', 'Rating': 8.8},
        {'Year': 2021, 'Game': 'Halo Infinite', 'Sales_Million': 20, 'Genre': 'Shooter', 'Platform': 'Xbox', 'Rating': 8.3},
        
        # 2022
        {'Year': 2022, 'Game': 'Elden Ring', 'Sales_Million': 25, 'Genre': 'RPG', 'Platform': 'Multi', 'Rating': 9.6},
        {'Year': 2022, 'Game': 'God of War Ragnarök', 'Sales_Million': 15, 'Genre': 'Action', 'Platform': 'PS5', 'Rating': 9.4},
        {'Year': 2022, 'Game': 'Pokémon Scarlet/Violet', 'Sales_Million': 25, 'Genre': 'RPG', 'Platform': 'Switch', 'Rating': 7.5},
        {'Year': 2022, 'Game': 'Call of Duty MW2', 'Sales_Million': 28, 'Genre': 'Shooter', 'Platform': 'Multi', 'Rating': 8.0},
        
        # 2023
        {'Year': 2023, 'Game': 'Baldurs Gate 3', 'Sales_Million': 15, 'Genre': 'RPG', 'Platform': 'Multi', 'Rating': 9.6},
        {'Year': 2023, 'Game': 'Zelda: TOTK', 'Sales_Million': 20, 'Genre': 'Adventure', 'Platform': 'Switch', 'Rating': 9.5},
        {'Year': 2023, 'Game': 'Spider-Man 2', 'Sales_Million': 11, 'Genre': 'Action', 'Platform': 'PS5', 'Rating': 9.0},
        {'Year': 2023, 'Game': 'Starfield', 'Sales_Million': 12, 'Genre': 'RPG', 'Platform': 'Xbox', 'Rating': 8.3},
        
        # 2024
        {'Year': 2024, 'Game': 'Helldivers 2', 'Sales_Million': 12, 'Genre': 'Shooter', 'Platform': 'Multi', 'Rating': 8.8},
        {'Year': 2024, 'Game': 'Black Myth Wukong', 'Sales_Million': 18, 'Genre': 'Action', 'Platform': 'Multi', 'Rating': 8.9},
        {'Year': 2024, 'Game': 'Dragon Age Veilguard', 'Sales_Million': 5, 'Genre': 'RPG', 'Platform': 'Multi', 'Rating': 8.0},
        {'Year': 2024, 'Game': 'Final Fantasy VII Rebirth', 'Sales_Million': 7, 'Genre': 'RPG', 'Platform': 'PS5', 'Rating': 9.2},
        
        # 2025
        {'Year': 2025, 'Game': 'GTA VI', 'Sales_Million': 30, 'Genre': 'Action', 'Platform': 'Multi', 'Rating': 9.5},
        {'Year': 2025, 'Game': 'Death Stranding 2', 'Sales_Million': 8, 'Genre': 'Action', 'Platform': 'PS5', 'Rating': 8.5},
        {'Year': 2025, 'Game': 'Monster Hunter Wilds', 'Sales_Million': 10, 'Genre': 'Action', 'Platform': 'Multi', 'Rating': 8.7},
        {'Year': 2025, 'Game': 'Assassins Creed Shadows', 'Sales_Million': 12, 'Genre': 'Action', 'Platform': 'Multi', 'Rating': 8.3},
    ]
    
    df = pd.DataFrame(games_data)
    print(f"✓ Created dataset with {len(df)} games across {df['Year'].nunique()} years")
    return df

# Creating the dataset
df = create_games_dataset()

print("\n" + "=" * 80)
print("DATASET PREVIEW")
print("=" * 80)
print(df.head(10))

print("\n" + "=" * 80)
print("DATASET STATISTICS")
print("=" * 80)
print(df.describe())

print("\nGenre Distribution:")
print(df['Genre'].value_counts())

# 1. Top Games by Sales (Bar Chart)
def plot_top_games_by_sales(df, top_n=15):
    """Plot top games by total sales"""
    print(f"\n Creating Top {top_n} Games by Sales chart")
    
    top_games = df.nlargest(top_n, 'Sales_Million')
    
    fig, ax = plt.subplots(figsize=(14, 10), facecolor='#0f1419')
    ax.set_facecolor('#1a1d24')
    
    
    genres = top_games['Genre'].unique()
    colors_map = plt.cm.Set3(np.linspace(0, 1, len(genres)))
    genre_colors = {genre: colors_map[i] for i, genre in enumerate(genres)}
    colors = [genre_colors[genre] for genre in top_games['Genre']]
    
    bars = ax.barh(range(len(top_games)), top_games['Sales_Million'], 
                   color=colors, alpha=0.8, edgecolor='black', linewidth=1.5)
    
    ax.set_yticks(range(len(top_games)))
    ax.set_yticklabels(top_games['Game'], fontsize=11, color='white')
    ax.set_xlabel('Sales (Million Units)', fontsize=13, fontweight='bold', color='white')
    ax.set_title(f'Top {top_n} Best-Selling Games (2015-2025)', 
                 fontsize=16, fontweight='bold', color='white', pad=20)
    ax.tick_params(colors='white')
    ax.grid(axis='x', alpha=0.3, linestyle='--')
    
   
    for i, (bar, val, year) in enumerate(zip(bars, top_games['Sales_Million'], top_games['Year'])):
        ax.text(val + 1, i, f'{val}M ({year})', 
               va='center', fontweight='bold', fontsize=10, color='white')
    
   
    legend_elements = [mpatches.Patch(facecolor=genre_colors[genre], 
                                     edgecolor='black', label=genre)
                      for genre in genres]
    ax.legend(handles=legend_elements, loc='lower right', fontsize=10,
             framealpha=0.9, facecolor='#1a1d24', edgecolor='white')
    
    plt.tight_layout()
    plt.show()

# 2. Sales Trend Over Years
def plot_sales_trend_by_year(df):
    """Plot total game sales trend by year"""
    print("\n📈 Creating Sales Trend by Year chart...")
    
    yearly_sales = df.groupby('Year')['Sales_Million'].sum().reset_index()
    yearly_count = df.groupby('Year').size().reset_index(name='Game_Count')
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), facecolor='#0f1419')
    
    # Total Sales
    ax1.set_facecolor('#1a1d24')
    ax1.plot(yearly_sales['Year'], yearly_sales['Sales_Million'], 
            marker='o', linewidth=3, markersize=10, color='#10b981', 
            markerfacecolor='#34d399', markeredgecolor='black', markeredgewidth=2)
    ax1.fill_between(yearly_sales['Year'], yearly_sales['Sales_Million'], 
                     alpha=0.3, color='#10b981')
    ax1.set_xlabel('Year', fontsize=13, fontweight='bold', color='white')
    ax1.set_ylabel('Total Sales (Million Units)', fontsize=13, fontweight='bold', color='white')
    ax1.set_title('Total Game Sales by Year (2015-2025)', 
                 fontsize=16, fontweight='bold', color='white', pad=15)
    ax1.grid(True, alpha=0.3, linestyle='--')
    ax1.tick_params(colors='white', labelsize=11)
    
    
    for x, y in zip(yearly_sales['Year'], yearly_sales['Sales_Million']):
        ax1.text(x, y + 10, f'{y:.0f}M', ha='center', fontweight='bold', 
                fontsize=10, color='white',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='black', alpha=0.7))
    
    
    ax2.set_facecolor('#1a1d24')
    bars = ax2.bar(yearly_count['Year'], yearly_count['Game_Count'], 
                   color='#3b82f6', alpha=0.8, edgecolor='black', linewidth=1.5)
    ax2.set_xlabel('Year', fontsize=13, fontweight='bold', color='white')
    ax2.set_ylabel('Number of Popular Games', fontsize=13, fontweight='bold', color='white')
    ax2.set_title('Number of Popular Game Releases by Year', 
                 fontsize=16, fontweight='bold', color='white', pad=15)
    ax2.grid(True, alpha=0.3, linestyle='--', axis='y')
    ax2.tick_params(colors='white', labelsize=11)
    
    
    for bar in bars:
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                f'{int(height)}', ha='center', va='bottom', 
                fontweight='bold', fontsize=11, color='white')
    
    plt.tight_layout()
    plt.show()

# 3. Genre Distribution
def plot_genre_analysis(df):
    """Analyze and visualize genre distribution"""
    print("\n🎯 Creating Genre Analysis charts...")
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8), facecolor='#0f1419')
    
    # Genre Sales Distribution (Pie Chart)
    ax1.set_facecolor('#1a1d24')
    genre_sales = df.groupby('Genre')['Sales_Million'].sum().sort_values(ascending=False)
    colors = plt.cm.Set3(np.linspace(0, 1, len(genre_sales)))
    
    wedges, texts, autotexts = ax1.pie(genre_sales.values, labels=genre_sales.index, 
                                        autopct='%1.1f%%', startangle=90, colors=colors,
                                        textprops={'color': 'white', 'fontweight': 'bold', 'fontsize': 11},
                                        wedgeprops={'edgecolor': 'black', 'linewidth': 2})
    ax1.set_title('Game Sales Distribution by Genre', 
                 fontsize=16, fontweight='bold', color='white', pad=20)
    
    
    ax2.set_facecolor('#1a1d24')
    genre_rating = df.groupby('Genre')['Rating'].mean().sort_values(ascending=False)
    
    bars = ax2.barh(range(len(genre_rating)), genre_rating.values, 
                    color=colors[:len(genre_rating)], alpha=0.8, 
                    edgecolor='black', linewidth=1.5)
    ax2.set_yticks(range(len(genre_rating)))
    ax2.set_yticklabels(genre_rating.index, fontsize=11, color='white')
    ax2.set_xlabel('Average Rating', fontsize=13, fontweight='bold', color='white')
    ax2.set_title('Average Game Rating by Genre', 
                 fontsize=16, fontweight='bold', color='white', pad=15)
    ax2.set_xlim(0, 10)
    ax2.grid(axis='x', alpha=0.3, linestyle='--')
    ax2.tick_params(colors='white')
    
   
    for i, (bar, val) in enumerate(zip(bars, genre_rating.values)):
        ax2.text(val + 0.1, i, f'{val:.2f}', va='center', 
                fontweight='bold', fontsize=11, color='white')
    
    plt.tight_layout()
    plt.show()




print("\n" + "=" * 80)
print("GENERATING VISUALIZATIONS")
print("=" * 80)

plot_top_games_by_sales(df, top_n=15)
plot_sales_trend_by_year(df)
plot_genre_analysis(df)


print("\n" + "=" * 80)
print("KEY INSIGHTS")
print("=" * 80)

print("\n Top 5 Best-Selling Games:")
print("-" * 80)
top5 = df.nlargest(5, 'Sales_Million')
for idx, game in top5.iterrows():
    print(f"{game['Game']:30s} | {game['Sales_Million']:3.0f}M units | {game['Year']} | {game['Genre']}")

print("\n Sales Trends:")
print("-" * 80)
yearly_sales = df.groupby('Year')['Sales_Million'].sum()
peak_year = yearly_sales.idxmax()
print(f"Peak Sales Year: {peak_year} with {yearly_sales[peak_year]:.0f}M units")
print(f"Total Sales (2015-2025): {df['Sales_Million'].sum():.0f}M units")
print(f"Average Sales per Game: {df['Sales_Million'].mean():.1f}M units")

print("\n Genre Analysis:")
print("-" * 80)
genre_stats = df.groupby('Genre').agg({
    'Sales_Million': ['sum', 'mean'],
    'Rating': 'mean',
    'Game': 'count'
}).round(2)
print(genre_stats)

print("\n Platform Performance:")
print("-" * 80)
platform_stats = df.groupby('Platform').agg({
    'Sales_Million': 'sum',
    'Game': 'count'
}).sort_values('Sales_Million', ascending=False)
print(platform_stats)

```

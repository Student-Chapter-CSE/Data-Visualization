```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
categories = np.repeat(['A', 'B', 'C', 'D'], 80)
values = np.hstack([
    np.random.normal(10, 2, 80),
    np.random.normal(15, 3, 80),
    np.random.normal(20, 2.5, 80),
    np.random.normal(25, 1.5, 80)
])

unique_cats = sorted(set(categories))
x_positions = {cat: i for i, cat in enumerate(unique_cats)}
x = np.array([x_positions[c] for c in categories])

jitter_strength = 0.15
x_jittered = x + np.random.uniform(-jitter_strength, jitter_strength, size=len(x))

plt.figure(figsize=(8, 6))
plt.scatter(x_jittered, values, s=25, alpha=0.7, color="#4C72B0", edgecolor="white", linewidth=0.5)

plt.xticks(range(len(unique_cats)), unique_cats, fontsize=12)
plt.xlabel("Category", fontsize=13)
plt.ylabel("Value", fontsize=13)
plt.title("Swarm Plot (Matplotlib Only)", fontsize=15, pad=10)

plt.grid(alpha=0.2, linestyle='--')
plt.tight_layout()
plt.show()
```

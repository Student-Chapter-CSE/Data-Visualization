```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud

print("=" * 70)
print("WORD CLOUD VISUALIZATION")
print("=" * 70)
print("\nBigger words = More frequent | Smaller words = Less frequent")
print("=" * 70)

# Product Reviews
sample_text_1 = """
Amazing product! The quality is excellent and the price is great.
I love this product. It's fantastic and works perfectly.
Highly recommend this product to everyone. Quality is outstanding.
The design is beautiful and the performance is excellent.
Great value for money. The product exceeded my expectations.
Fast delivery and great customer service. Very satisfied with my purchase.
The product is durable and well-made. I'm very happy with it.
Excellent quality and great features. Worth every penny.
Love the color and the design. It's exactly what I needed.
Perfect product. Highly recommend to anyone looking for quality.
"""

# Sample text - Book Review
sample_text_2 = """
This book is absolutely amazing! The story is captivating and the characters
are well-developed. The author's writing style is engaging and beautiful.
The plot twists kept me reading until the very end. I couldn't put it down.
The character development is excellent and the dialogue feels natural.
This is one of the best books I've read this year. Highly recommend it!
The storytelling is masterful and the pacing is perfect throughout.
Every chapter brings something new and exciting. A must-read for everyone!
The emotional depth of the characters is remarkable. Truly a wonderful book.
"""

print("\n Sample Text 1: Product Reviews")
print("-" * 70)
print(sample_text_1[:150] + "...")

print("\n Sample Text 2: Book Review")
print("-" * 70)
print(sample_text_2[:150] + "...")

def create_wordcloud(text, title="Word Cloud", colormap='viridis', background='white'):
    print(f"\nðŸŽ¨ Creating word cloud: {title}...")
    
    wordcloud = WordCloud(width=1200, height=600, background_color=background,
                         colormap=colormap, max_words=100).generate(text)
    
    fig, ax = plt.subplots(figsize=(15, 8), facecolor=background)
    ax.imshow(wordcloud, interpolation='bilinear')
    ax.axis('off')
    
    title_color = 'white' if background == '#1a1d24' else 'black'
    ax.set_title(title, fontsize=20, fontweight='bold', pad=20, color=title_color)
    
    plt.tight_layout()
    plt.show()
    print(f"âœ“ Word cloud created!")

print("\n" + "=" * 70)
print("GENERATING WORD CLOUDS")
print("=" * 70)

create_wordcloud(sample_text_1, title="Product Reviews", colormap='viridis')

create_wordcloud(sample_text_2, title="Book Review", colormap='cool', background='#1a1d24')



```

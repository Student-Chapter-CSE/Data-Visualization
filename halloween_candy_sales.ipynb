```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle
import matplotlib.patches as mpatches
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
# Halloween candy sales data (in millions of dollars)
candy_types = [
    'Chocolate Bars',
    'Candy Corn',
    'Gummy Worms',
    'Lollipops',
    'Sour Candies',
    'Caramel Apples',
    'Peanut Butter Cups',
    'Licorice'
]

sales_millions = [45.2, 32.8, 28.5, 25.3, 22.7, 18.9, 52.6, 15.4]

# Spooky Halloween colors
spooky_colors = [
    '#FF6B35',  
    '#FF9F1C',  
    '#9B59B6', 
    '#E74C3C',  
    '#27AE60',  
    '#8B4513',  
    '#FFA500',  
    '#000000'   
]
```

```python
# Sort data by sales for better visualization
sorted_indices = np.argsort(sales_millions)[::-1]
sorted_candies = [candy_types[i] for i in sorted_indices]
sorted_sales = [sales_millions[i] for i in sorted_indices]
sorted_colors = [spooky_colors[i] for i in sorted_indices]
```

```python
# Create the main Halloween-themed visualization
fig, ax = plt.subplots(figsize=(14, 8))

# Set spooky background
fig.patch.set_facecolor('#1a1a2e')  
ax.set_facecolor('#16213e')  

# Create horizontal bars
bars = ax.barh(sorted_candies, sorted_sales, 
               color=sorted_colors, 
               alpha=0.9, 
               edgecolor='#FFA500', 
               linewidth=2.5)

# Add pumpkin emoji/icons at the end of each bar
for i, (candy, sales) in enumerate(zip(sorted_candies, sorted_sales)):
    # Add sales value
    ax.text(sales + 1.5, i, f'${sales:.1f}M', 
            va='center', ha='left', 
            fontsize=11, 
            fontweight='bold',
            color='#FFA500')
    
    # Add pumpkin icons 
    pumpkin_count = int(sales / 10)  
    pumpkin_str = '' * min(pumpkin_count, 5) 
    ax.text(-3, i, pumpkin_str, 
            va='center', ha='right', 
            fontsize=14)

# Customize the chart with Halloween theme
ax.set_xlabel('Sales (Millions USD)', 
              fontsize=14, 
              fontweight='bold', 
              color='#FFA500')
ax.set_ylabel('Candy Type', 
              fontsize=14, 
              fontweight='bold', 
              color='#FFA500')
ax.set_title(' Halloween Candy Sales 2024 \n"Spooky Sweet Treats Analysis"', 
             fontsize=18, 
             fontweight='bold', 
             color='#FF6B35',
             pad=20)

# Style the ticks
ax.tick_params(colors='#FFA500', which='both')
ax.spines['bottom'].set_color('#FFA500')
ax.spines['top'].set_color('#FFA500')
ax.spines['left'].set_color('#FFA500')
ax.spines['right'].set_color('#FFA500')

# Set tick label colors
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_color('#FFA500')
    label.set_fontweight('bold')

# Add grid with spooky style
ax.grid(axis='x', alpha=0.2, linestyle='--', color='#FFA500')
ax.set_axisbelow(True)

# Add total sales annotation with spooky styling
total_sales = sum(sorted_sales)
textstr = f' Total Sales \n${total_sales:.1f}M'
props = dict(boxstyle='round,pad=0.8', 
             facecolor='#FF6B35', 
             edgecolor='#FFA500',
             linewidth=3,
             alpha=0.9)
ax.text(0.98, 0.97, textstr, 
        transform=ax.transAxes, 
        fontsize=13,
        fontweight='bold',
        verticalalignment='top', 
        horizontalalignment='right', 
        bbox=props,
        color='#1a1a2e')

# Add Halloween decorations
decoration_text = '     '
ax.text(0.5, -0.12, decoration_text, 
        transform=ax.transAxes,
        fontsize=20,
        ha='center',
        va='top')

# Add legend for pumpkin icons
legend_text = ' = $10M in sales'
ax.text(0.02, 0.02, legend_text,
        transform=ax.transAxes,
        fontsize=10,
        fontweight='bold',
        color='#FFA500',
        bbox=dict(boxstyle='round,pad=0.5', 
                  facecolor='#1a1a2e', 
                  edgecolor='#FFA500',
                  linewidth=2,
                  alpha=0.8))

plt.tight_layout()
plt.show()
```

```python
# Print spooky statistics
print("\n" + "="*50)
print(" HALLOWEEN CANDY SALES ANALYSIS 2024 ")
print("="*50)
print(f"\n Total Candy Types Analyzed: {len(candy_types)}")
print(f" Total Market Sales: ${total_sales:.2f} Million")
print(f" Average Sales per Candy: ${total_sales/len(candy_types):.2f} Million")
print(f"\n Top Seller: {sorted_candies[0]} (${sorted_sales[0]:.1f}M)")
print(f" Runner Up: {sorted_candies[1]} (${sorted_sales[1]:.1f}M)")
print(f" Third Place: {sorted_candies[2]} (${sorted_sales[2]:.1f}M)")
print(f"\n Least Popular: {sorted_candies[-1]} (${sorted_sales[-1]:.1f}M)")

# Calculate market share
print("\n MARKET SHARE BREAKDOWN:")
print("-" * 50)
for candy, sales in zip(sorted_candies, sorted_sales):
    share = (sales / total_sales) * 100
    bar = '' * int(share / 5)  # One pumpkin per 5% share
    print(f"{candy:.<25} {share:>5.1f}% {bar}")

print("\n" + "="*50)
print(" Happy Halloween! Enjoy your treats! ")
print("="*50 + "\n")
```

```python
# Bonus: Create a spooky pie chart
fig, ax = plt.subplots(figsize=(12, 8))
fig.patch.set_facecolor('#1a1a2e')

# Create pie chart with Halloween colors
wedges, texts, autotexts = ax.pie(
    sorted_sales[:5],  # Top 5 candies
    labels=sorted_candies[:5],
    colors=sorted_colors[:5],
    autopct='%1.1f%%',
    startangle=90,
    explode=[0.05, 0.02, 0.02, 0.02, 0.02],
    shadow=True,
    textprops={'color': '#FFA500', 'fontweight': 'bold', 'fontsize': 11}
)

# Style the percentage text
for autotext in autotexts:
    autotext.set_color('#1a1a2e')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(12)

ax.set_title(' Top 5 Halloween Candy Market Share ', 
             fontsize=18, 
             fontweight='bold',
             color='#FF6B35',
             pad=20)

# Add Halloween decorations
decoration = '       '
plt.figtext(0.5, 0.02, decoration, ha='center', fontsize=16)

plt.tight_layout()
plt.show()
```

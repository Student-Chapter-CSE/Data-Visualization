```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Player': ['Virat Kohli', 'Shikhar Dhawan', 'Rohit Sharma', 'David Warner', 'Suresh Raina', 'AB de Villiers', 'MS Dhoni'],
    'Runs': [7263, 6617, 6259, 6397, 5528, 5162, 5082],
    'Matches': [237, 217, 243, 176, 205, 184, 250],
}

df = pd.DataFrame(data)
df['Average'] = (df['Runs'] / df['Matches']).round(2)
df = df.sort_values(by='Runs', ascending=False)

plt.figure(figsize=(10,6))
bars = plt.bar(df['Player'], df['Runs'], color='skyblue', edgecolor='black')
plt.plot(df['Player'], df['Average']*100, color='red', marker='o', label='Average  100')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 100, yval, ha='center', va='bottom', fontsize=9)

plt.title('Top IPL Run Scorers and Their Batting Averages', fontsize=14, fontweight='bold')
plt.xlabel('Player', fontsize=12)
plt.ylabel('Total Runs', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
